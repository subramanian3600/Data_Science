```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("train_new.csv" ,index_col='Unnamed: 0')
df1
```

```python
df2=pd.read_csv("test_new.csv",index_col='Unnamed: 0')
df2
```

```python
missing_cols = set(df1.columns ) - set(df2.columns )
for c in missing_cols:
    df2[c] = 0
df2
```

```python
merged=pd.concat([df1,df2],ignore_index=True)
merged
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test=train_test_split(merged,shuffle=False,random_state=10,train_size=0.68)
```

```python
x=X_train[['Pclass','Sex', 'Age', 'Parch', 'Fare','Survived']]
y=X_test[['Survived']]
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train[['Pclass','Sex', 'Age', 'Parch', 'Fare']],X_train[['Survived']])
lr.predict(X_test[['Pclass','Sex', 'Age', 'Parch', 'Fare']])
```

```python
lr.score(X_train[['Pclass','Sex', 'Age', 'Parch', 'Fare']],X_train[['Survived']])
```
