# HR STATISTICS FOR THE EMPLOYERS LEFT THE COMPANY

```python
import pandas as pd
hr_df=pd.read_csv("hr.csv")
hr_df
```

```python
sal=pd.get_dummies(hr_df.salary)
sal
```

```python
hr_df1=hr_df.drop(['salary'],axis='columns')
merged=pd.concat([hr_df1,sal],axis='columns')
merged
```

```python
left=hr_df[hr_df.left==1]
left
```

```python
retained=hr_df[hr_df.left==0]
retained
```

```python
a=merged.groupby(hr_df.left).mean()
a
```

```python
fin=merged.drop(['number_project','last_evaluation','Work_accident','time_spend_company'],axis='columns')
fin
```

```python
dept=pd.get_dummies(hr_df.Department)
dept
merge_dept=pd.concat([merged,dept],axis='columns')
merge_dept
final=merge_dept.drop(['Department'],axis='columns')
final
```

```python
x=final[['satisfaction_level','average_montly_hours','promotion_last_5years','high','low','medium']]
#'IT','RandD','accounting','hr','management','marketing','product_mng','sales','support','technical'
y=final.left
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
log.predict(x_test)
```

## accuracy of model

```python
log.score(x_test,y_test)
```
