```python
amazon_sellers=pd.read_csv("amazon_sellers.csv")
amazon_sellers
```

```python
amazon_sellers.drop('Name',axis='columns')
```

```python
amazon_sellers.info()
#there are no null obj
```

# Yearly Fiction and Non-Fiction Movies User Rating Visuvalation

```python
ax=sns.lineplot(x='Year',y='User Rating',data=amazon_sellers,hue='Genre',ci=False)
```
