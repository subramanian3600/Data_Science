PER CAPITA INCOME PREDICTION EXERCISE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
pci_df=pd.read_csv("pci.csv")
pci_df
```

```python
x=pci_df.year
y=pci_df.per_capita_income
plt.scatter(x,y,color='g',marker='+')
plt.xlabel("YEAR")
plt.ylabel("PER CAPITA INCOME")
plt.title("PER CAPITA INCOME")
```

```python
reg_pci=linear_model.LinearRegression()
reg_pci.fit(pci_df[['year']],pci_df.per_capita_income)
```

```python
new=pd.read_csv("pci_new.csv")
new
```

```python
b=reg_pci.predict(new[['year']])
b
```

```python
new['per_capita_income']=b
new
```

```python
final=pd.concat([pci_df,new],ignore_index=True)
final
```
