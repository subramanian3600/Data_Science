# GDP INDIA 2020

GDP from 1961 to 2019:

```python
import pandas as pd
df=pd.read_csv("gdp.csv")
df
```

A Graph Representing GDP Before COVID19:

```python
x=df['Year']
y=df['GDP growth (annual %)']
import matplotlib.pyplot as plt
plt.plot(x,y,label='GDP',color='g')
plt.xlabel("year")
plt.ylabel("GDP")
plt.title("GDP India")
plt.legend(loc='lower left')
```

```python
df20=pd.DataFrame({
    "Year":[2020],
    "GDP growth (annual %)": [-23]
})
df20
newdf=pd.concat([df,df20],ignore_index=True)
newdf
```

A Graph Representing GDP After COVID19:

```python
s=newdf['Year']
b=newdf['GDP growth (annual %)']
import matplotlib.pyplot as plt
plt.plot(s,b ,label='GDP',color='r')
plt.xlabel("year")
plt.ylabel("GDP")
plt.title("GDP India")
plt.legend()
```
